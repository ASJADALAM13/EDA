### Description:
This dataset contains information on 1K+ Amazon products, including their ratings, reviews, and other details.

## Features:

```python
product_id: Unique identifier for each product

product_name: Name of the product

category: Category of the product

discounted_price: Discounted price of the product 

actual_price: Actual price of the product 

discount_percentage: Percentage of discount for the product 
                                           
rating: Rating of the product (1-5) 
                                           
rating_count: Number of people who voted for the Amazon rating 
                                  
about_product: Description about the product 

user_id: ID of the user who wrote the review
                                  
user_name: Name of the user who wrote the review 
                                  
review_id: ID of the user review 
                                  
review_title: Short review 
                                  
review_content: Long review
                                  
img_link: Image link of the product 
                                  
product_link: Official website link of the product
```

### changing dtype of some columns

```python
##removing the '₹' symbol
df['discounted_price'] = df['discounted_price'].str.replace('₹','')

##removing the ','
df['discounted_price'] = df['discounted_price'].str.replace(',','')
```

```python
df['discounted_price']
```

```python
df['discounted_price'] = df['discounted_price'].astype('float')
```

```python
df['discounted_price']
```

```python
df['actual_price']
```

```python
df['discount_percentage'] = df['discount_percentage'].str.replace("%","")
df['discount_percentage'] = df['discount_percentage'].astype(int)
```

```python
df['discount_percentage']
```

```python
## rating col
```

```python
df['rating']
```

```python
df['rating'] = pd.to_numeric(df["rating"],errors="coerce")
```

```python
df['rating']
```

```python
df['rating_count']
```

```python
df['rating_count'] = df['rating_count'].str.replace(",",'')
```

```python
df['rating_count'] = pd.to_numeric(df['rating_count'],errors='coerce')
```

```python
df['rating_count']
```

```python
df['rating_count'].isnull().sum()
```

```python
## REPLACE NULL VALUE WITH MODE
```

```python
df['rating_count'] = df['rating_count'].fillna(df['rating_count'].mode()[0])
```

```python
df['rating_count'].isnull().sum()
```

## 3. What is the distribution of discounted prices vs. actual prices?

```python
dist =["discounted_price","actual_price"]
fig,ax =plt.subplots(nrows=2,ncols=1,figsize=(12,12))
for i,col in enumerate(dist):
    sns.histplot(x=col,data=df,ax=ax[i],kde=True)
    ax[i].set_title(col)
plt.tight_layout()
plt.show()
```

## insight:
#### discount price :
#### discount price range from 0-80000.
#### 0-10000 price had more no of product counts.


####   actual price :
#### actual price range 0-140000.
#### 0-20000 price has more no of product count

## 6. What are the most popular product keywords?

```python
# Group by 'product_name' and count the occurrences to find the most popular product
popular_product_keyword = df.groupby('product_name').size().reset_index(name='count')

# Sort the products by their count in descending order (most popular products first)
popular_product_keyword = popular_product_keyword.sort_values(by='count', ascending=False)

# Display the top products
print(popular_product_keyword.head())
```
